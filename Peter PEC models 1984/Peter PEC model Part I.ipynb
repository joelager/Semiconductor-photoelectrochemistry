## Overview of the Peter approach for the Part I paper

1.  A p-type semiconductor is assumed: under illumination the generation rate $g$ corresponds to electrons (minority carrriers generated in the space charge region (SCR) plus one diffusion length of the bulk (Gärtner model). 
2.  In contrast to approach of Reichman (see other notebook), recombination in the space charge region is not considered. Instead, recombination at a surface state is added to the model. At flat band, this state is empty and positively charged and can capture electrons which then can recombine with the majority carrier holes. As the voltage drop across the SCR is increased, the surface state will drop below the Fermi level and become filled and neutral. In its filled state, it cannot capture electrons so this recombination pathway will be turned off and the photocurrent will approach the generation rate $g$. The current onset occurs when the surfaces states are effectively turned off by changing their occupancy with bias. This is why the current onset is shifted from the flat band potential.  
3. A current continuity balance is used to derive the photocurrent. Unless the surface state is fully occupied, there must be an additional flux of holes from the bulk $j_p$ to meet the electrons at the surface state which is equal, at steady state, to the recombination current for electrons. The photocurrent is thus just the generation rate minus the excess hole current: $j_{photo} = g - j_p$  
4. There is a calculation of the effect of surface charge state on the total voltage drop including both the double layer and the SCR. For large surface state densities, the voltage drop across the double layer can be large when the surface state is charged. 
5.  There is useful information in the transients (this analysis) and impedance (other papers in the series) data.  Consider a p-type semiconductor just below the steady state onset.  In a transient experiment,there is a quick pulse of cathodic current but after a short time holes figure out that there are electrons available at the surface  and diffuse there to meet them.  Another way to think about it is that the surface electrons are there as soon as the light is on but the holes need time to diffuse there to meet them.  



* $\Delta \phi$ is the voltage drop in the semiconductor;
* The current to the solution is just proportional to the electron concentration $n_c$ in the CB at the surface. First order kinetics is used with rate constant $k_0$  
* p-type semiconductor with properties like GaP 
* One surface state (donor) 0.3 eV above VBM. This state is empty at flat band and becomes fills when the downward band bending exceeds 0.3 V. It is positively charged when empty and becomes neutral when it captures and electron.   
* Only electrons can flow to the electrolyte - there is no majority carrier current. But holes can flow to the surface state, recombining with electrons there.  

## Relationship between photocurrent and surface state occupancy
Generation rate $g$ is that of Gärtner:

$$g = I_0 \frac{1 - \exp(-\alpha W)}{1 + \alpha L_n} \tag{1}$$

where $\alpha$ is absorption coefficient, $I_0$ is illumination in current density units, $L_n$ is the bulk electron diffusion length, and $W$ is the width of the space charge region:

$$W(\Delta \phi) = \left( \frac{2 \Delta \phi \epsilon \epsilon_0}{q N_s} \right) ^{1/2} \tag{2}$$

where $\epsilon$ is the relative permittivity (dielectric constant), $q$ is the elementary charge, and $N_s$ is surface state density. 

The change in the population of electrons in the surface state (where they are available to recombine with holes) compared to the equilibrium is

$$\Delta n_s(t) = N_s \Delta f(t) \tag{3a}$$

where $\Delta f(t)$ is the change in surface state occupation 

$$ \Delta f(t) = f(t) - f_0 \tag{3b}$$

and where $f_0$ is the surface state occupancy in the dark defined by Fermi-Dirac statistics:  

$$f_0 = \left[ 1 + \exp\left( \frac {E_s - q \Delta \phi}{k_B T} \right)  \right]^{-1} \tag{3c}$$

Excess hole flux is then proportional to surface concentration of holes, concentration of surface states, and the change in their occupancy:

$$j_p(t) = k_p p_s N_s \Delta f(t) \tag{4} $$

where the surface concentration of holes decreases with increasing voltage drop in the SCR:

$$p_s = p_0 \exp \left( \frac{-q\Delta \phi}{k_B T} \right) \tag{5}$$

and rate constant $k_p$ for hole capture is related to capture cross section and hole thermal velocity $k_p = \sigma_p v_p$ 

Time rate of change of the surface electron population $n_c$ balances generation, surface state recombination, and cathodic current:

$$\frac{dn_c}{dt} = g - k_n N_s [1 - f(t)]n_c - k_0 n_c \tag{6}$$
Three notes about Eq. (6):  
* Recombination turns off if the traps are filled $f(t) = 1$  
* Electron flux is not necessarily the same as the hole flux except in steady state  
* Electrons cannot get to the solution from the surface. This is reasonable for a sufficiently wide bandgap and a surface state located close to the VBM  

The photocurrent due to reactions of the surface electrons with redox species in the solution is now 

$$j_{photo}(t) = g - j_p(t) \tag{7}$$

It is done this way, with the majority carriers, to anticipate the non-steady state solutions later.  In steady state, you should get the same answer, with $j_{photo} = k_0 n_c$. 

Referring back to Eq. (4), $\Delta f(t)$ defines the time dependence of $j_{photo}$. So its time dependence is determined by the electron and hole fluxes to the surface state 

$$N_s \frac{df(t)}{dt} = k_n N_s [1 - f(t)]) n_c(t) - k_p p_s N_s \Delta f(t) \tag{8}$$

## Steady state solution

Solving Eqs. (6) and (8) in the steady state, $dn_s/dt = 0$ and $df(t)/dt = 0$ yields the steady state surface state occupancy $f_{\infty}$ and steady state electron concentration.  From setting Eq. (6) = 0:

$$n_c = \frac{g}{k_n N_s (1 - f_{\infty}) + k_0} \tag{9}$$

Substituting Eq. (9) into Eq. (8) yields a quadratic equation for the steady state surface state occupancy whose solution is  

$$f_{\infty} = \frac{-B - \sqrt{B^2 - 4 A C}}{2A} \tag{10}$$ 

with

$$A = k_n k_p p_s N_s^2 \tag{11a}$$

$$B = -[k_n N_s g + k_n k_p p_s N_s^2 (1+f_0) + k_0 k_p p_s N_s] \tag{11b}$$

$$C = k_n N_s g + k_n k_p p_s N_s^2 f_0 + k_0 k_p p_s N_s f_0 \tag{11c}$$

Finally, using Eqs. (3b) and (4) the steady state photocurrent is 

$$j_{photo}(\infty) = g - k_p p_s N_s(f_{\infty} - f_0) \tag{12}$$

## Surface states and the potential distribution

Charged (empty) surface states will create a potential drop across the Helmholtz layer; this will diminish as the states are filled and become neutral.  
> *In the most extreme case, the Fermi level will be pinned by the surface state and the changes in potential will largely appear across the Helmholtz layer*

For surface states which are positive when vacant (donors), potential drop across the double layer is derived from $C V = Q$

$$\Delta V_H = \frac{q N_s (1-f_0)}{C_H} + \textrm{const} \tag{13}$$

Zero of potential is taken as the flat band potential ($\Delta \phi = 0$)

At a given potential, light will change the surface state occupancy as electrons go there to recombine but the effect turns out to be small in steady state as will be shown below

$$\delta \Delta \phi = \frac{q N_s (f_{\infty} -f_0)}{C_H} \tag{14}$$

## Transient solution

Recalling that  
$$j_{photo}(t) = g(t) - j_p(t)$$

$$ j_p(t) = k_p p_s N_s \Delta f(t)$$ 

$$\Delta f(t) = f(t) -f_0$$

one can see that determining $\Delta f(t)$ will yield the transient photocurrent. From Eq. (8) and assuming that the surface electron responds instantaneously to changes in illumination such that the steady state solution in Eq. (9) can be used for $n_c(t)$ yields 

$$N_s \frac{d \Delta f(t)}{dt} = N_s \frac{d}{dt} [f(t) - f_0] = \frac{k_n N_s [1 - f(t)]g(t)}{k_n N_s (1 - f(t))+ k_0} - k_p p_s N_s \Delta f(t)  \tag{15}$$

Rewriting Eq. 15 yields the following ODE for $\Delta f(t)$

$$\frac{d \Delta f(t)}{dt} =  \frac{k_n [1 - f_0 - \Delta f(t)]g(t)}{k_n N_s [1 - f_0 - \Delta f(t)]+ k_0} - k_p p_s \Delta f(t)  $$

Equation (15) for the surface state occupancy has 2 terms.  The first is increase due to light induced electron current.  This turns on instantly. The second is decrease in occupancy due to the hole flux to the surface state.  This depends a lot on bias. 

In the paper, Peter and co-workers solve Eq. (15) using a recursion relationship. Here, it will be solved with `solve_ivp` from the `scipy` library. With $\Delta f(t)= f(t) - f_0$ known, then $j_p(t) = k_p p_s N_s \Delta f(t)$ and $j_{photo} = g(t) - j_p(t) $


An approximate solution is also derived.  At low light intensities or high surface state densities, $\Delta f(t)$ will be small so that $f(t) \approx f_0$

In this approximation, the time dependence of the surface occupancy is 

$$\Delta f(t) = \frac{g}{k_p p_s N_s [1+k_0/N_s(1-f_0)]} \left(1-\exp(-t/\tau) \right) \tag{16}$$

With the time dependence being entirely determined by the surface hole concentration

$$\tau_0(\Delta \phi) = \frac{1}{k_p \, p_s(\Delta \phi)}\tag {17}$$

This yields the transient photocurrent ($g$ is the generation rate at $t =0$)

$$j_{photo}(t) = g \left[  1 - \frac{1-\exp(-t/\tau)}{1+k_0/N_s(1-f_0)} \right] \tag{18}$$

Using the same approximation, steady state photocurrent is 

$$j_{photo}(\infty) = g \left[  1 - \frac{1}{1+k_0/N_s(1-f_0)} \right] \tag{19}$$

In this approximation, if you substract the steady-state current $j{photo}(\infty)$ from the time dependent value and also from the initial value, a simple form is generated:

$$\frac{j_{photo}(t)-j_{photo}(\infty)}{j_{photo}(0)-j_{photo}(\infty)} = \exp(t/\tau) \tag{20}$$

Still...  
> *Equation 20 illustrates the limiting exponential behaviour which is expected when surface-state filling can be neglected, but in general Eq. (15) must be solved numerically to obtain the transient photocurrent response.*    

Which is what is done below

# Material, surface, and illumination classes

```python
# Make classes for material, illumination, and surface states

class Material():
    """
    Default is conditions of Figure 2
    These are not changed within the paper
    """
    def __init__(self, 
                α=3.3e4*cm**-1,      # 300 nm absorption depth
                p0=8.0e16*cm**-3,    # Same as the doping
                NA=8e16*cm**-3,      # p-type
                Ln=7e-6*cm,          # 70 nm
                ϵ=11,
                T=300*kelvin):
        self.α = α
        self.p0 = p0
        self.NA = NA    # Might be ok as it is inside a class
        self.Ln = Ln
        self.ϵ = ϵ
        self.T = T
               
class Illumination():
    """
    Illumination in units of current density
    This is changed for the different cases  
    E.g. from 1e12 to 1e16 in Figure 3
    """
    def __init__(self, 
                I0=1e14*cm**-2*sec**-1):   # Is about 16 μA cm^-2
        self.I0 = I0
    

class Surface():
    """
    Default is Figure 3 and 4 parameters
    These are changed for the different cases
    Figure 2 varies Ns from 1e11 to 1e14
    """
    def __init__(self, 
                kn=1e-7*cm**3*sec**-1,   # vthermal = 1e7, trap cross section 1e-14
                kp=1e-8*cm**3*sec**-1,   # vthermal = 1e7, trap cross section 1e-15
                k0=1e4*cm*sec**-1,       # Note GaP-like semiconductor, this is close to VBM
                Ns=1e13*cm**-2,          # Used in Figure 3 and 4 
                Es=0.3*eV,               # Close to VBM
                C_Helmholtz=2e-5*farad*cm**-2):         
        self.kn = kn
        self.kp = kp
        self.k0 = k0
        self.Es = Es
        self.Ns = Ns
        self.C_Helmholtz = C_Helmholtz
        
# Instantiate the classes with default parameters   
mat1 = Material()
surf1 = Surface()
illum1 = Illumination()
```

# Steady state analysis

```python
# Make a big class for the steady state model

class Steady_State():
    
    def g(self, Δϕ, mat=Material, illum=Illumination):
        """
        Eq. 1
        Gartner generation rate
        """
        term1 = np.exp(-mat.α*self.W(Δϕ, mat=mat))
        term2 = 1 + mat.α*mat.Ln
        return illum.I0*(1 - term1/term2)
    
    def W(self, Δϕ, mat=Material):
        """
        Eq. 2
        Width of space charge layer
        """
        if Δϕ > 0:
            W = ((2*Δϕ*mat.ϵ*ϵ0)/(q*mat.NA))**0.5
        else:
            W = 0
        return W

    def f0(self, Δϕ, mat=Material, surf=Surface):
        """
        Equation 3c
        Surface state occupancy in the dark and 
        the surface concentation of holes (the majority carriers)
        Simple Fermi - Dirac statistics for a discrete state 
        """
        term1 = np.exp((surf.Es - q*Δϕ)/(kB*mat.T))
        return (1 + term1)**-1

    def p_surface(self, Δϕ, mat=Material):
        """
        Equation 5
        Surface concentration of holes
        """
        return mat.p0*np.exp((-q*Δϕ)/(kB*mat.T))
    
    def f_inf(self, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 10
        Surface state occupany under illumination in steady state
        """
        # Make local variables 
        kn = surf.kn
        kp = surf.kp
        k0 = surf.k0
        Ns = surf.Ns
        p_surf = self.p_surface(Δϕ, mat=mat)
        g = self.g(Δϕ, mat=mat, illum=illum) 
        f0 = self.f0(Δϕ, mat=mat, surf=surf)
    
        # Carefully construct the terms of the quadratic
        A = kn*kp*p_surf*Ns**2
        B1 = kn*Ns*g 
        B2 = kn*kp*p_surf*Ns**2*(1 + f0)
        B3 = k0*kp*p_surf*Ns
        B = -(B1 + B2 + B3)
        C1 = kn*Ns*g
        C2 = kn*kp*p_surf*Ns**2*f0
        C3 = k0*kp*p_surf*Ns*f0
        C = C1 + C2 + C3
        return (-B - np.sqrt(B**2 - 4*A*C))/(2*A)
    
    def ncSteadyState(self, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 9
        electron concentration at the surface
        Numerator is photogeneration, demoninator is rate to the empty trap 
        and to the solution
        """
        term1 = self.g(Δϕ, mat=mat, illum=illum) 
        term2 = surf.kn*surf.Ns*(1 - self.f_inf(Δϕ, mat=mat, surf=surf, illum=illum))
        term3 = surf.k0
        return term1/(term2 + term3)
    
    def jPhotoSteadyState(self, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 12
        Photocurrent is electron generation minus the 
        light-induced change of the rate of surface holes going 
        to the trap (to meet the electrons...)  
        """
        generation = self.g(Δϕ, mat=mat, illum=illum) 
        f_infinity = self.f_inf(Δϕ, mat=mat, surf=surf, illum=illum)
        f_zero = self.f0(Δϕ, mat=mat, surf=surf)
        recombination = surf.kp*self.p_surface(Δϕ, mat=mat)*surf.Ns*(f_infinity - f_zero)
        return generation - recombination

# Instantiate the steady state model    
model = Steady_State()

# Depletion of surface holes
print(f'{model.p_surface(0.0, mat=mat1)/cm**-3:.2e}')

# Check of electron concentration
print(f'{model.ncSteadyState(0, mat=mat1, surf=surf1, illum=illum1):.2e}')
```

# Analysis of surface charging

Graphs above are vs. V drop in the SC layer $\Delta \phi$.  
> But experimentally, the applied voltage can drop across the SCR and across the Helmholtz layer.  And changes in surface state occupancy will change the V drop across the Helmholtz layer assuming that its capacitance does not change. 

For this p-type case it is assumed that the surface states are + when vacant and neutral when they are filled (e.g. they are donors). In this case, see above, the capacitance is defined by $CV = Q$

```python
# Make a class for the Helmholtz layer

class Helmholtz_layer(Steady_State):    # Make it a child of the Steady_State class
    
    def ΔV_Helmholtz_dark(self, Δϕ, mat=Material, surf=Surface):
        """
        Equation 13
        Voltage drop across Helmholtz layer in the dark
        Is is q*ns when the states are empty
        Charge decreases as states fill with electons
        Note the reference to the other class
        """
        term1 = q*surf.Ns*(1 - self.f0(Δϕ, mat=mat, surf=surf))
        return term1/surf.C_Helmholtz

    def δΔV_Helmholtz_light(self, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 14
        Change in the surface charging due to light
        """
        term1 = q*surf.Ns*(self.f_inf(Δϕ, mat=mat, surf=surf, illum=illum) \
                           - self.f0(Δϕ, mat=mat, surf=surf))
        return term1/surf.C_Helmholtz
    
model2 = Helmholtz_layer()
```

# Transient analysis

Will compare the numerical and approximate solutions of Equation (15).  
Will model only a single on/off pulse although other pulse shapes are certainly possible.  

```python
class Transient(Steady_State):
    """
    Will inherit all of the methods from the Steady_State class
    """

    def τ_0(self, Δϕ, mat=Material, surf=Surface, **kwargs):    # Note use of kwargs so I can overload it
        """
        Eqn 17
        Electron lifetime at surface
        """
        return (surf.kp*self.p_surface(Δϕ, mat=mat))**-1
    
    def jPhotoSteadyStateEqn19(self, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 19, 
        an approximation for the steady state photocurrent, ignores light effect on
          trap occupancy
        """
        generation = self.g(Δϕ, mat=mat, illum=illum) # Light just turns on
        term1 = surf.k0
        term2 = surf.kn*surf.Ns*(1 - self.f0(Δϕ, mat=mat, surf=surf))
    
        return generation*(1 - (1 + term1/term2)**-1) 
    
    def jPhotoTransEqn18(self, t, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        Equation 19, 
        an approximation for the transient photocurrent, ignores light effect on
          trap occupancy
        """
        generation = self.g(Δϕ, mat=mat, illum=illum) # Light just turns on
        τ = self.τ_0(Δϕ, mat=mat, surf=surf)
        term0 = np.exp(-t/τ)
        term1 = surf.k0
        term2 = surf.kn*surf.Ns*(1 - self.f0(Δϕ, mat=mat, surf=surf))
    
        return generation*(1 - (1 - term0)/(1 + term1/term2)) 
   
    def g_on_off(self, t, Δϕ, width=1*μs, mat=Material, illum=Illumination):
        """
        Function for a single light pulse
        """
        return np.heaviside(width - t, 1)*self.g(Δϕ, mat=mat, illum=illum)
    
    
    def IVP_Eqn15(self, t, Δf_dum, Δϕ, mat=Material, surf=Surface, illum=Illumination):
        """
        returns dΔf/dt to be used in solve_ivp
          hard coded for a pulse width of 4*τ
        """
        # Make local variables
        Δf = Δf_dum
        f_t = self.f0(Δϕ, mat=mat, surf=surf) + Δf
        kn = surf.kn
        kp = surf.kp
        k0 = surf.k0
        Ns = surf.Ns
        p_surf = self.p_surface(Δϕ, mat=mat)
    
        τ = self.τ_0(Δϕ, mat=mat, surf=surf)
        g_t = self.g_on_off(t, Δϕ, width=4*τ, mat=mat, illum=illum)
    
        f0 = self.f0(Δϕ, mat=mat, surf=surf)
        
        # Coded exactly from the paper
        term1 = kn*Ns*(1 - f_t)*g_t
        term2 = kn*Ns*(1 - f_t) + k0
        term3 = kp*p_surf*Ns*Δf
    
        return (term1/term2 - term3)/Ns    # So have to divide out Ns at the end
    
    def Eqn15_solve(self, Δϕ, mat=Material, surf=Surface, illum=Illumination, n_points=1000):
        """
        Calling the solver
        Will return time, Δf(t), j_photo(t), and g(t)
          Has an optional argument for the number of points to return
        """
        IVP_Eqn15_to_solve = lambda t, Δf: self.IVP_Eqn15(t, Δf, Δϕ, mat=mat, surf=surf, illum=illum)
        
        τ = self.τ_0(Δϕ, mat=mat, surf=surf)
        t_max = 8*τ
        t_eval = np.linspace(0, t_max, n_points)
        #print(τ/μs)
        
        # IVP_Eqn15_to_solve(0, 0) 

        sol = solve_ivp(IVP_Eqn15_to_solve, [0, t_max], [0],
               t_eval=t_eval, method='Radau')
        
        j_p = surf.kp*self.p_surface(Δϕ, mat=mat)*surf.Ns*sol.y[0]
        g_t = self.g_on_off(sol.t, Δϕ, width=4*τ, mat=mat, illum=illum)

        j_photo = g_t - j_p
        
        return np.array([sol.t, sol.y[0], j_photo, g_t])

model3 = Transient()
```
