## Overview of the Peter approach for the Part III paper

**Synopsis**

n-type case with a surface state and Gerischer kinetics from VB, CB, and surface state to solution.  
Thermal detrapping is considered in the surface state occupany  
Steady state and impendence analysis with a periodically varying illumination.



**Band diagram**
* n-type semiconductor is assumed with acceptor state located at $E_s$ in between conduction $E_c$ and valence $E_v$ band edges  
* DOS in the electrolyte is centred at $E_{redox}$ and has O (acceptor-like) and R (donor-like) states available with a characteristic width of the solvent reorganization $\lambda$.  
* As shown in the figure, electrons at the Fermi level (dashed) can reduce species in the electrolyte (O), noting there is a potential drop between the semiconductor and the electrolyte ($E_F$ in the semiconductor does not line up with $E_{redox}$ in the solution.  
* There is one surface state (acceptor) 0.4 eV below CBM at $E_s$. This state is empty at flat band and becomes fills when the downward band bending $\Delta \phi$ exceeds ~0.25 V when it is reached by the bulk Fermi level. It is positively charged when empty and becomes neutral when it captures an electron.

**Kinetic scheme** 
* There is generation $G$ and also an electron current (majority carriers) going to the surface to recombine with the holes, the surface state being occupied as drawn and thus happy to accept a hole.  
* Electrons can go into surface state with rate constant $k_n$ and can get out thermally.  
* Holes can go into the surface state with rate constant $k_n$ and get out therally.
* electrons and holes in the surface state can react forwards and backwards with $ox$ and $red$ in the solution with rate constants $\overleftarrow{k}_s$ and $\overrightarrow{k}_s$
* Electron and hole transfer from the bands in both directions is also included: $\overleftarrow{k}_c, \overrightarrow{k}_c, \overleftarrow{k}_v, \overrightarrow{k}_v$ 


**Currents in and out of the solution**.  
There are 6 of them
* Electrons in CB react with $\ce{Ox}$ in solution: 
 > $-n_s \overrightarrow{k}_c \, c_{ox}$. Cathodic current decreases with increasing band banding ($n_s$ decreases), does not depend on light.  
* Reduced species in solution donate electrons into the CB: 
 > $ \overleftarrow{k}_c \, c_{red}$. Anodic. Does not depend on band bending or light.
* Electrons in surface state with $\ce{Ox}$ in solution:
   > $- \overrightarrow{k}_s N_{ss} \, c_{ox} f_{ss}$. Cathodic current which depends on trap occupancy (which is weakly changed by light). Is the dominant contribution to the dark current and deceases with increasing $\Delta \phi$ as the trap is emptied. 
* Reduced species in solution donate electrons into the surface state (if it is unoccupied):  
 > $\overleftarrow{k}_s N_{ss}\, c_{red} (1-f_{ss})$  Anodic current which turns on when trap is emptied. Rate depends on overlap of the DOS of the reduced species with the surface state. 
* Electrons in VB react with oxidized species in solution:  
 > $-\overrightarrow{k}_v \, c_{ox}$. Cathodic and small, does not depend on band bending or light  
* Holes in VB react with reduced species in solution.   
 > $p_s \overleftarrow{k}_v c_{red}$. Anodic and small in the dark, increases with band bending and light. Is the dominant photocurrent due to better overlap of the solution DOS with the band. 

## Relationship between photocurrent and surface state occupancy

Basic idea is extension of Hall-Shockley-Read (HSR) approach for surface recombination to the electrolyte/semiconductor interface. In the dark, time dependence of surface state occupancy $f_{ss}$ is determined by the net rates from the two bands and the solution: 

$$ \frac{d f_{ss}}{dt} = k_p p_1(1-f_{ss}) - k_p p_s f_{ss} + k_n n_s (1-f_{ss}) -k_n n_1 f_{ss} - \overrightarrow{k}_s c_{ox} f_{ss} + \overleftarrow{k}_s c_{red} (1-f_{ss})\tag{A1}$$

Solving Equation (A1) in steady state yields 

$$f_{ss} = \frac{k_p p_1 + k_n n_s + \overleftarrow{k}_s c_{red}}
                 {k_p(p_1 + p_s) + k_n (n_1 + n_s) + \overrightarrow{k}_s c_{ox} + \overleftarrow{k}_s c_{red} } \tag{1}$$
                 
where $p_1$ and $n_1$ are

$$p_1 = N_v \exp \left\{ (E_c - E_s)/k_B T \right\} \tag{2a}$$

$$n_1 = N_v \exp \left\{ (E_s - E_c)/k_B T \right\} \tag{2b} $$

Surface states in under illumination from solving (A6) and (A5). Note that Eq. (8c) had a typo in the paper. Also note that $p_s$ is not in these equations.

$$f_{ss}^{\ast} = \frac{-b - \sqrt{b^2 - 4ac}}{2a} \tag{8a}$$

$$ a = -N_{ss} k_p \{k_n(n_1 + n_s) +\overrightarrow{k}_s c_{ox} + \overleftarrow{k}_s c_{red} \} \tag{8b}$$

$$b = +N_{ss} k_p (k_n n_s + \overleftarrow{k}_s c_{red} ) -
\overleftarrow{k}_v c_{red} \{k_p p_1 + \overrightarrow{k}_s c_{ox} + k_n (n_s + n_1)  \} - k_p (G + \overrightarrow{k}_v c_{ox} )\tag{8c}$$

$$ c = \overleftarrow{k}_v c_{red} (k_p p_1 + k_n n_s + \overleftarrow{k}_s c_{red}) \tag{8d}$$

Generation rate $G$ is that of Gärtner:

$$G = I_0 \frac{1 - \exp(-\alpha W)}{1 + \alpha L_n} \tag{9}$$

where $\alpha$ is absorption coefficient, $I_0$ is illumination in current density units, $L_n$ is the bulk electron diffusion length, and $W$ is the width of the space charge region:

$$W(\Delta \phi) = \left( \frac{2 \Delta \phi \epsilon \epsilon_0}{q N_A} \right) ^{1/2} \tag{Appendix}$$

where $\epsilon$ is the relative permittivity (dielectric constant), $q$ is the elementary charge, and $N_A$ is acceptor density. 



Flux balance under illumination. The generation current has to equal to the net flux of holes to the solution

$$G = k_p p_s^{\ast} N_{ss}f_{ss}^{\ast} - k_p p_s N_{ss} (1 - f_{ss}^{\ast}) + \overleftarrow{k}_v c_{red} p_s^{\ast} - \overrightarrow{k}_v c_{ox} \tag{A5}$$

Compare electron and hole currents in the dark. These have 4 terms for each band - in/out of band and in/out trap.

$$J_h = \overleftarrow{k}_v c_{red} p_s + k_p p_s N_{ss}f_{ss} - k_p p_1 N_{ss}(1 - f_{ss}) - \overrightarrow{k}_v c_{ox}  \tag{A1}$$

$$J_e = \overrightarrow{k}_c c_{ox} n_s + k_p n_s N_{ss} (1 - f_{ss}) - k_n n_1 N_{ss}f_{ss} - 
\overleftarrow{k}_c c_{red}  \tag{A2}$$

## Surface hole concentration

In light and dark for electrons and in dark for holes (added by JWA)

$$n_s = N_C \exp \left( \frac{E_{F, bulk} -E_c - q \Delta \phi}{kT} \right)  $$

$$p_s = N_V \exp \left( \frac{E_v - E_{F, bulk} + q \Delta \phi}{kT} \right)  $$

In light, from paper

$$p_s^{\ast} = \frac{G + k_p p_1 N_{ss}(1 - f_{ss}^{\ast}) + \overrightarrow{k}_v c_{ox} }{k_p N_{ss} f_{ss}^{\ast} + \overleftarrow{k}_v c_{red}} \tag{A6}$$

But this does not go to the dark case when $G = 0$. It may be that steady state in the dark needs to be at the rest potential. 

## Steady state currents

$$J_{cb} = -n_s \overrightarrow{k}_c c_{ox} + \overleftarrow{k}_c c_{red} \tag{A7}$$   

$$J_{vb} = p_s \overleftarrow{k}_v c_{red} - \overrightarrow{k}_v c_{ox} \tag{A8}$$

$$\overrightarrow{J}_{ss} = \overrightarrow{k}_{s} N_{ss} c_{ox} f_{ss}^{\ast} \tag{A9}$$

$$\overleftarrow{J}_{ss} = \overleftarrow{k}_{s} N_{ss} c_{red}(1- f_{ss}^{\ast}) \tag{A10}$$

## Surface states and the potential distribution

Charged (empty) surface states will create a potential drop across the Helmholtz layer; this will diminish as the states are filled and become neutral.  
> *In the most extreme case, the Fermi level will be pinned by the surface state and the changes in potential will largely appear across the Helmholtz layer*

For surface states which are positive when vacant (donors), potential drop across the double layer is derived from $C V = Q$

$$\Delta V_H = \frac{q N_s (1-f_{ss})}{C_H} + \textrm{const} \tag{7}$$

This effect is small if surface site density is below $10^{12}$ cm$^{-3}$ as discussed in first paper.

Zero of potential is taken as the flat band potential ($\Delta \phi = 0$)



## Transient solution

Sinusoidal excitation is assumed  

$$G(t) = G_0(1 + \sin \, \omega t)\tag{10}$$

Surface sites are 

$$ f_{ss}^{\ast}(t) = \left\{ 1 + \frac{G_0 k_p \tau_v}{\delta} 
\left(  \frac{\omega\cos \, \omega t}{\omega^2 + \tau_s^{-2}} - \frac{\sin \, \omega t}{\tau_s(\omega^2 + \tau_s^{-2})} \right)\right\} \tag{11}$$

with

$$ \tau_s^{-1} = k_n(n_s + n_1) + \overleftarrow{k}_s c_{red} + \overrightarrow{k}_s c_{ox} \tag{12}$$

Normalized photocurrent

$$\frac{\tilde{J}_{photo}}{\tilde{G}} = \text{Re}\left( \frac{\tilde{J}_{photo}}{\tilde{G}}  \right) + 
\text{Im}  \left( \frac{\tilde{J}_{photo}}{\tilde{G}}   \right)  \tag{13}$$

where 

$$ \text{Re}\left( \frac{\tilde{J}_{photo}}{\tilde{G}}  \right) = 1 - \frac{\beta}{\omega^2 \tau_s^2 + 1} \tag{14a} $$

$$ \text{Im}\left( \frac{\tilde{J}_{photo}}{\tilde{G}}  \right) =  \frac{\beta \omega \tau_s^2}{\omega^2 \tau_s^2 + 1} \tag{14b} $$

$\beta$ is a rate constant

$$\beta = \frac{N_{ss} k_n (n_s + n_1) k_p f_{ss,0}^{\ast}}{\overleftarrow{k}_v c_{red} + k_p f_{ss,0}^{\ast} N_{ss} } \tag{15}$$

In low frequency limit

$$\left(\frac{\tilde{J}_{photo}}{\tilde{G}}\right)_{\omega \to 0} \approx 1 - \beta \tau_s \tag{16}$$

such that $\tilde{J}_{photo}/\tilde{G}$ plotted in complex plane should be semicircles of diameter $\beta \tau_s$. The frequency at which the plot is a maximum, $\omega_m$ allows direct determination of the rate at which surface states relax

$$\omega_m = \tau_s^{-1}$$

## Rate constants

```python
# Class for the rate constants for charge exchange with the solution

class Rate_Constants():
    """
    There are six rate constants for interacting with the solution: 
      forward and reverse for the CB, surface state, and VB
      none of them depend on potential
    """
    def kv_rev(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 4a
        electron from reduced species to VB (if there are holes there).
        units of cm^4 sec^-1
        """
        pre1 = sol.σ_red_ν_p_δ*((kB*mat.T)/(np.pi*sol.λ_red))**0.5
        expon1 = (mat.E_v - sol.E_red)**2
        expon2 = 4*sol.λ_red*kB*mat.T
        return pre1*np.exp(-expon1/expon2)
        
    def kc_for(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 4b 
        electron from CB to oxidized species 
        units of cm^4 sec^-1
        """
        pre1 = sol.σ_ox_ν_n_δ*((kB*mat.T)/(np.pi*sol.λ_ox))**0.5
        expon1 = (sol.E_ox - mat.E_c)**2
        expon2 = 4*sol.λ_ox*kB*mat.T
        return pre1*np.exp(-expon1/expon2)
    
    def kv_for(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 5a - carrier injection
        electron from (filled) valence band to oxidized species
        units of cm*sec^-1
        """
        Z = sol.Z
        expon1 = (sol.E_ox - mat.E_v)**2
        expon2 = 4*sol.λ_ox*kB*mat.T
        return Z*np.exp(-expon1/expon2)
        
    def kc_rev(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 5b - carrier injection
        Electron from reduced species to the CB
        units of cm*sec^-1
        """
        Z = sol.Z
        expon1 = (mat.E_c - sol.E_red)**2
        expon2 = 4*sol.λ_red*kB*mat.T
        return Z*np.exp(-expon1/expon2)
    
    def ks_for(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 6a
        This is back reaction, 
        electron from filled trap to reduce species in solution
        units of cm^3*s^-1
        """
        pre1 = sol.κ_ss*sol.Z*sol.δ**2
        pre2 = ((kB*mat.T)/(np.pi*sol.λ_ox))**0.5
        expon1 = (sol.E_ox - surf.E_s)**2      # 1.85 eV - 1.0 eV
        expon2 = 4*sol.λ_ox*kB*mat.T
        return pre1*pre2*np.exp(-expon1/expon2)
        #return expon1/expon2
    
    def ks_rev(self, mat=Material, surf=Surface, sol=Solution):
        """
        Equation 6b
        electron from solution to the empty trap generating an 
        oxidation current
        units of cm^3*s^-1
        """
        pre1 = sol.κ_ss*sol.Z*sol.δ**2
        pre2 = ((kB*mat.T)/(np.pi*sol.λ_red))**0.5
        expon1 = (sol.E_red - surf.E_s)**2      # 0.15 - 1 
        expon2 = 4*sol.λ_red*kB*mat.T
        return pre1*pre2*np.exp(-expon1/expon2)
        #return expon1/expon2

rate_constants = Rate_Constants()
```

## Surface state occupancy and carrier statistics

```python
class Steady_State(Rate_Constants):
    """
    In all the function defs will add **kwargs so that I can
      send extra named arguments as will happen when I make 
      dictionaries for grouping the parameters
    All units are SI but the cgs units for reporting are noted
    """
    
    def f_ss(self, Δϕ, mat=Material, surf=Surface, sol=Solution, **kwargs):
        """
        Equation 1 and A4
        Surface state occupancy in the dark
        unitless
        """
        # use local variables
        kp = surf.kp
        kn = surf.kn
        p1 = self.p1(mat=mat, surf=surf)
        n1 = self.n1(mat=mat, surf=surf)
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        c_ox = sol.c_ox
        c_red = sol.c_red
        p_s = self.p_s(Δϕ, mat=mat)
        n_s = self.n_s(Δϕ, mat=mat)
        
        term1 = kp*p1 + kn*n_s + ks_rev*c_red
        term2 = kp*(p1 + p_s) + kn*(n1 + n_s)
        term3 = ks_for*c_ox + ks_rev*c_red
        
        return term1/(term2 + term3)

    def p_s(self, Δϕ, mat=Material, **kwargs):
        """
        surface hole concentration in the dark
        Not written out explicitly in paper
        """
        return mat.N_v*np.exp((mat.E_v - mat.EF_bulk + q*Δϕ)/(kB*mat.T))
    
    def n_s(self, Δϕ, mat=Material, **kwargs):
        """
        surface electron concentration, light and dark
        Not written out explicitly in paper
        """
        return mat.N_c*np.exp((mat.EF_bulk - mat.E_c - q*Δϕ)/(kB*mat.T))
    
    def p1(self, mat=Material, surf=Surface, **kwargs):
        """
        Equation 2a
        These take into account thermal transitions to the bands
        Units of cm^-3
        """
        p1 = mat.N_v*np.exp((mat.E_v - surf.E_s)/(kB*mat.T))
        return p1
 
                           
    def n1(self, mat=Material, surf=Surface, **kwargs):
        """
        Equation 2b
        Units of cm^-3
        """
        n1 = mat.N_c*np.exp((surf.E_s - mat.E_c)/(kB*mat.T))
        return n1
    

    def V_H(self, mat=Material, surf=Surface):
        """
        Equation 7
        """
        pass
    
    def f_ss_star(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation 8
        Note that this will not work if N_ss=0
        But it does with with no illumination
        Unitless
        """
        # use local variables
        Ns = surf.Ns
        kp = surf.kp
        kn = surf.kn
        p1 = self.p1(mat=mat, surf=surf)
        n1 = self.n1(mat=mat, surf=surf)
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        kv_for = self.kv_for(mat=mat, surf=surf, sol=sol)
        c_ox = sol.c_ox
        c_red = sol.c_red
        p_s = self.p_s(Δϕ, mat=mat)
        n_s = self.n_s(Δϕ, mat=mat)
        G = self.g(Δϕ, mat=mat, illum=illum)
        
        # From Mathematica solver
        # a seems to agree note minus sign
        aM = -kp*Ns*(c_red*ks_rev + c_ox*ks_for + kn*(n1 + n_s))
        bM = G*kp + c_red**2*ks_rev*kv_rev + \
              c_ox*c_red*ks_for*kv_rev + c_ox*kp*kv_for + \
              c_red*kn*kv_rev*n1 + \
              c_red*kn*kv_rev*n_s - \
              c_red*kp*ks_rev*Ns - \
              kn*kp*n_s*Ns + \
              c_red*kp*kv_rev*p1
        
        # Factor this to find the typo
        bM2 = Ns*kp*(c_red*ks_rev + kn*n_s) \
             -kp*(G + c_ox*kv_for) \
              -kv_rev*c_red*(c_red*ks_rev + c_ox*ks_for + kn*n1 + kn*n_s + kp*p1)
        
        cM = c_red*kv_rev*(c_red*ks_rev + kn*n_s + kp*p1)
        
        # agrees - note that this will be zero if no surface states
        a = -Ns*kp*(kn*(n1 + n_s) + ks_for*c_ox + ks_rev*c_red)
        
        # Corrected typo in Eqn 8b
        b = +Ns*kp*(kn*n_s +ks_rev*c_red) - \
             kv_rev*c_red*(kp*p1 + ks_for*c_ox + ks_rev*c_red + kn*(n_s + n1)) -\
             kp*(G + kv_for*c_ox)
        
        # agrees
        c = kv_rev*c_red*(kp*p1 + kn*n_s + ks_rev*c_red)
        
        sol_M = (-bM2 - np.sqrt(bM2**2 - 4*aM*cM))/(2*aM)
        
        sol_Peter_fixed = (-b - np.sqrt(b**2 - 4*a*c))/(2*a)
        
        return sol_Peter_fixed
    
    def p_s_star(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation A6
        Surface concentration of holes with light
        Might need another equation for the no surface state case
        Does not seem to converge to the dark case in the dark
        """
        # use local variables
        Ns = surf.Ns
        kp = surf.kp
        kn = surf.kn
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        kv_for = self.kv_for(mat=mat, surf=surf, sol=sol)
        p1 = self.p1(mat=mat, surf=surf)
        c_ox = sol.c_ox
        c_red = sol.c_red
        G = self.g(Δϕ, mat=mat, illum=illum)
        f_ss_star = self.f_ss_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        term1 = G + kp*p1*Ns*(1 - f_ss_star) + kv_for*c_ox
        term2 = kp*Ns*f_ss_star + kv_rev*c_red
        
        return term1/term2
        
    
    def g(self, Δϕ, mat=Material, illum=Illumination, **kwargs):
        """
        Eq. 9
        Gartner generation rate
        Units of amp*cm^-2
        """
        term1 = np.exp(-mat.α*self.W(Δϕ, mat=mat))
        term2 = 1 + mat.α*mat.L_p
        return illum.I0*(1 - term1/term2)
   
    #@np.vectorize - not easy to make this work within a class
    def W(self, Δϕ, mat=Material, **kwargs):
        """
        Is not written out explicitly in paper
        Width of space charge layer
        Units of length e.g. nm
        """
        if Δϕ > 0:
            W = ((2*Δϕ*mat.ϵ*ϵ0)/(q*mat.N_d))**0.5
        else:
            W = 0
        return W

# Instantiate for testsing
stats = Steady_State()
```

## Steady state currents

```python
 class Steady_State_Currents(Steady_State):
    
    def J_cb_for(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        Equation A7, cathodic part
        does  not depend on light
        """
        kc_rev = self.kc_rev(mat=mat, surf=surf, sol=sol)
        kc_for = self.kc_for(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        c_ox = sol.c_ox
        n_s = self.n_s(Δϕ, mat=mat)
        return -n_s*kc_for*c_ox

    
    def J_cb_rev(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        Equation A7, anodic part
        does  not depend on light
        """
        kc_rev = self.kc_rev(mat=mat, surf=surf, sol=sol)
        kc_for = self.kc_for(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        c_ox = sol.c_ox
        n_s = self.n_s(Δϕ, mat=mat)
        return kc_rev*c_red
    
    def J_vb_for(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        Equation A8, cathodic part
        Tiny and negative in the dark
        Does not depend on lighti
        """
        kv_for = self.kv_for(mat=mat, surf=surf, sol=sol)
        c_ox = sol.c_ox        
        return - kv_for*c_ox
    
    def J_vb_rev(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation A8, anodic part
        Tiny and negative in the dark (very few holes), positive and important in the light
        This is the main anodic current under illumination 
        """
        p_s_star = self.p_s_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        return p_s_star*kv_rev*c_red 

    
    def Js_for(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation A9
        Occupied surface state reducing Ox in solution
        Depends on light but dependence is weak
        """
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        Ns = surf.Ns
        c_ox = sol.c_ox
        f_ss_star = self.f_ss_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        return -ks_for*Ns*c_ox*f_ss_star
    
    def Js_rev(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation A10, reduced species reacting with empty surface state
        Another source of anodic current if the surface state is not filled
        """
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        Ns = surf.Ns
        c_red = sol.c_red
        f_ss_star = self.f_ss_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        return ks_rev*Ns*c_red*(1 - f_ss_star)
    
    def J_vb(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Convenience function for the sum 
        """
        J_vb = self.J_vb_for(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum) + \
               self.J_vb_rev(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum) 
        return J_vb
    
    def J_total(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Add Equations A7-A10
        """
        J_cb = self.J_cb_for(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum) + \
               self.J_cb_rev(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        J_vb = self.J_vb_for(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum) + \
               self.J_vb_rev(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum) 
        Js_for = self.Js_for(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        Js_rev = self.Js_rev(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        return J_cb + J_vb + Js_for + Js_rev
    
    def J_photo(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Sum the anodic photocurrents - this might not be right
        """
        p_s_star = self.p_s_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        
        J_vb_photo = p_s_star*kv_rev*c_red
        Js_rev = self.Js_rev(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        return J_vb_photo + Js_rev

    # now the dark versions
    # cb currents are the same in the light or dark 
    
    def J_cb_dark(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        CB current does not depend on light
        """
        J_cb = self.J_cb_for(Δϕ, mat=mat, surf=surf, sol=sol) + \
               self.J_cb_rev(Δϕ, mat=mat, surf=surf, sol=sol)
        return J_cb
   
    def J_vb_dark(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        Equation A8, for holes in vb
        Reverse current depends on light
        """
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        kv_for = self.kv_for(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        c_ox = sol.c_ox
        Ns = surf.Ns
        p_s = self.p_s(Δϕ, mat=mat)
        f_ss = self.f_ss(Δϕ, mat=mat, surf=surf, sol=sol)

        return p_s*kv_rev*c_red - kv_for*c_ox
    
    def J_ss_dark(self, Δϕ, 
                  mat=Material, surf=Surface, 
                  sol=Solution, **kwargs):
        """
        Equations A9 and A10, for the surface state
        """
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        c_ox = sol.c_ox
        Ns = surf.Ns
        f_ss = self.f_ss(Δϕ, mat=mat, surf=surf, sol=sol)

        return -ks_for*Ns*c_ox*f_ss + ks_rev*Ns*c_red*(1 - f_ss)
    
    
    def J_total_dark(self, Δϕ, 
                       mat=Material, surf=Surface, 
                       sol=Solution, **kwargs):
        """
        total dark current
        CB current does not depend on light
        """
        return self.J_cb_dark(Δϕ, mat=mat, surf=surf, sol=sol) + \
               self.J_vb_dark(Δϕ, mat=mat, surf=surf, sol=sol) + \
               self.J_ss_dark(Δϕ, mat=mat, surf=surf, sol=sol)
    
    

# Instantiate the steady state model    
model = Steady_State_Currents()




print(f'p1: {model.p1(mat=mat1, surf=surf1, sol=sol1)/cm**-3:.2e}')
print(f'n1: {model.n1(mat=mat1, surf=surf1, sol=sol1)/cm**-3:.2e}')

print()
print(f'p_s: {model.p_s(0*volt, mat=mat1, surf=surf1, sol=sol1)/cm**-3:.2e}')
print(f'n_s: {model.n_s(0*volt, mat=mat1, surf=surf1, sol=sol1)/cm**-3:.2e}')


"""

print(model.f_ss(0.4, mat=mat1, surf=surf1, sol=sol1), 
      model.f_ss_star(0.4, mat=mat1, surf=surf1, sol=sol1,  illum=illum1))


print(model.J_vb_dark(0.2, mat=mat1, surf=surf1, sol=sol1), 
      model.J_vb(0.2, mat=mat1, surf=surf1, sol=sol1,  illum=illum1))

# model.p_s_star(0.2, mat=mat1, surf=surf1, sol=sol1, illum=illum1)
"""
print()
```

# Transient code

```python
class Transient_Currents(Steady_State):
    
    def β(self, Δϕ, ω, 
                  mat=Material, surf=Surface, 
                  sol=Solution, illum=Illumination):
        """
        Equation 15 for a first order state
        """
        Ns = surf.Ns
        kn = surf.kn
        n_s = self.n_s(Δϕ, mat=mat)
        n1 = self.n1(mat=mat, surf=surf)
        kp = surf.kp
        f_ss_star_zero = self.f_ss_star(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        kv_rev = self.kv_rev(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
 
        term1 = Ns*kn*(n_s + n1)*kp*f_ss_star_zero 
        term2 = kv_rev*c_red + kp*f_ss_star_zero*Ns
        
        return term1/term2
    
    def τ_s(self, Δϕ, 
            mat=Material, surf=Surface, 
            sol=Solution, **kwargs):
        """
        Equation 12
        Surface lifetime, does not depend on frequency or light
        """
        kn = surf.kn
        n_s = self.n_s(Δϕ, mat=mat)
        n1 = self.n1(mat=mat, surf=surf)
        ks_rev = self.ks_rev(mat=mat, surf=surf, sol=sol)
        ks_for = self.ks_for(mat=mat, surf=surf, sol=sol)
        c_red = sol.c_red
        c_ox = sol.c_ox
        
        term1 = kn*(n_s + n1)
        term2 = ks_rev*c_red + ks_for*c_ox
        
        return (term1 + term2)**-1
    
    def J_trans_norm_real(self, Δϕ, ω, 
                          mat=Material, surf=Surface, 
                          sol=Solution, illum=Illumination):
        """
        Equation 13
        """
        β = self.β(Δϕ, ω, mat=mat, surf=surf, sol=sol, illum=illum)
        τ_s = self.τ_s(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        term1 = β*τ_s
        term2 = ω**2*τ_s**2 + 1
        return 1 - term1/term2
    
    def J_trans_norm_imag(self, Δϕ, ω, 
                          mat=Material, surf=Surface, 
                          sol=Solution, illum=Illumination):
        """
        Equation 13
        """
        β = self.β(Δϕ, ω, mat=mat, surf=surf, sol=sol, illum=illum)
        τ_s = self.τ_s(Δϕ, mat=mat, surf=surf, sol=sol, illum=illum)
        
        term1 = β*ω*τ_s**2
        term2 = ω**2*τ_s**2 + 1
        return term1/term2
    
    # For the Bode plots
    
    def J_trans_mag(self, Δϕ, ω, 
                          mat=Material, surf=Surface, 
                          sol=Solution, illum=Illumination):
        J_trans_norm_real = self.J_trans_norm_real(Δϕ, ω,
                                                  mat=mat, surf=surf, 
                                                  sol=sol, illum=illum)
        J_trans_norm_imag = self.J_trans_norm_imag(Δϕ, ω,
                                                  mat=mat, surf=surf, 
                                                  sol=sol, illum=illum)
        return np.sqrt(J_trans_norm_real**2 + J_trans_norm_imag**2)
    
        # For the Bode plots
    
    def J_trans_phase(self, Δϕ, ω, 
                          mat=Material, surf=Surface, 
                          sol=Solution, illum=Illumination):
        J_trans_norm_real = self.J_trans_norm_real(Δϕ, ω,
                                                  mat=mat, surf=surf, 
                                                  sol=sol, illum=illum)
        J_trans_norm_imag = self.J_trans_norm_imag(Δϕ, ω,
                                                  mat=mat, surf=surf, 
                                                  sol=sol, illum=illum)
        return np.arctan2(J_trans_norm_imag, J_trans_norm_real)

    
trans = Transient_Currents()

trans.β(0.25, 1000, **default_case)
    
trans.τ_s(0.25, **default_case)

trans.J_trans_norm_imag(0.25, 1000, **default_case)

```
